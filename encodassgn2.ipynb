```python
#Sol1..

Ordinal Encoding:
    Here, we have give a unique rank to unique categoriacal data.
    
Example: Consider a categorical feature "Education Level" with categories: High School, Bachelor's,
Master's, PhD. Ordinal encoding might assign High School=0, Bachelor's=1, Master's=2, PhD=3, reflecting
the order of education levels.

Use : Ordinal encoding is chosen when the categorical data has a clear ranking or order, such as 
 educational levels, ratings (low, medium, high), or sizes (small, medium, large).
    
Label Encoding:
    categorical features encoded aphabetical and ranked in alphabetical order..
    
Example: Suppose you have a categorical feature "Color" with three categories: Red, Green, Blue. 
Label encoding might assign Red=0, Green=1, Blue=2.

Use : Label encoding is suitable when the categorical feature doesn't have an inherent order or 
hierarchy among its categories. For instance, when encoding categories like different types of fruits
or city names.
    
```

```python
#Sol2..

Target guided ordinal encoding:

It is a technique used to encode categorical variable based on relationship with target variable..
we replace each category in categorical variable with numearivcal value based mean and median of target variable..

When we use it:
Target Guided Ordinal Encoding can be useful when you have categorical features that might have a strong
correlation with the target variable.

For instance, in real estate, neighborhoods often have a significant
impact on house prices. Encoding neighborhoods based on their relationship with house prices can help the 
model understand and leverage this relationship more effectively
```

```python
#Sol3..

Covariance is a statistical measure that describes the relationship between two random variables. 
It measures how much two variables change together. In essence, it indicates whether an increase in 
one variable corresponds to an increase or decrease in the other.

Mathematically, covariance between two variables, X and Y, can be calculated using the following formula:

Cov(x,y)=1/nâˆ‘(x-p)(y-q)

Cov(X,Y) is the covariance between variables X and Y.

whwre p and q are the mean of x and y data respectavely..
```

```python
#Sol4..

from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Sample dataset
data = {
    'Color': ['red', 'green', 'blue', 'red', 'blue'],
    'Size': ['small', 'medium', 'large', 'medium', 'small'],
    'Material': ['wood', 'metal', 'plastic', 'wood', 'metal']
}

# Convert the dictionary to a pandas DataFrame
df = pd.DataFrame(data)

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode categorical variables
for col in df.columns:
    df[col + '_encoded'] = label_encoder.fit_transform(df[col])

# Display the encoded dataset
df
```

```python
import pandas as pd

# Sample dataset (replace this with your actual dataset)
data = {
    'Temperature': [25, 28, 22, 30, 26],
    'Humidity': [60, 55, 70, 45, 50],
    'Weather Condition': ['Sunny', 'Cloudy', 'Rainy', 'Sunny', 'Cloudy'],
    'Wind Direction': ['North', 'South', 'East', 'West', 'North']
}

# Convert the dictionary to a pandas DataFrame
df = pd.DataFrame(data)

# Calculate the covariance between Temperature and Humidity
cov_temp_humidity = df['Temperature'].cov(df['Humidity'])

print(f"Covariance between Temperature and Humidity: {cov_temp_humidity}")
```

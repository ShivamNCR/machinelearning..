# KNN Regression

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import make_regression

X,y=make_regression(n_samples=1000,
    n_features=2,
    noise=10,
    random_state=42)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
KNregressor=KNeighborsRegressor(n_neighbors=5,algorithm='auto')
```

```python
KNregressor.fit(X_train,y_train)
```

```python
y_pred=KNregressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_pred,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'n_neighbors':[1,2,3,4,5],
           'weights':['uniform', 'distance'],
           'algorithm':['auto', 'ball_tree', 'kd_tree'],
           'leaf_size':[20,30],
           'metric': ['euclidean', 'manhattan', 'minkowski']
           }
```

```python
grid=GridSearchCV(KNeighborsRegressor(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
ypred4=grid.predict(X_test)
```

```python
print(r2_score(ypred4,y_test))
```

```python
#Sol1.

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_excel('flight_price.xlsx')
df.head(2)
```

```python
df.shape
```

```python
#Sol2..

# Plot histogram
plt.figure(figsize=(6, 3))
plt.hist(df['Price'], bins=30, color='orange', edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Sol3..

df.describe()
```

```python
#Sol4..

# Plot boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(df, x='Airline', y='Price', palette='Set3')
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.grid(True)  # Add grid lines for better visualization
plt.show()
```

```python
#Sol 13...

data['Price'].unique()
```

```python
char_remove=['+',',','$']
for item in char_remove:
    data['Price']=data['Price'].str.replace(item,'')
    
```

```python
data['Price'].unique()
```

```python
data['Price']=data['Price'].replace('Everyone',np.nan)
```

```python
data['Price']=pd.to_numeric(data['Price'],errors='coerce')
```

```python
data['Price'].dtype
```

```python
data['Category']=data['Category'].replace('1.9',np.nan)
```

```python
data['Category'].value_counts()
```

```python
average_prices = data.groupby('Category')['Price'].mean()
```

```python
average_prices
```

```python
# Create a bar graph
plt.figure(figsize=(10, 6))
average_prices.plot(kind='bar', color='skyblue')
plt.title('bar graph with average prices of each category')
```

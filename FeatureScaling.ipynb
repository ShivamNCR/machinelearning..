```python
#Sol3.

Principal component analysis, or PCA, is a dimensionality reduction method that is often used to reduce
the dimensionality of large data sets, by transforming a large set of variables into a smaller one that 
still contains most of the information in the large set.

#EX.
Using PCA can help identify correlations between data points, such as whether there is a correlation between 
consumption of foods like frozen fish and crisp bread in Nordic countries. Principal component analysis today
is one of the most popular multivariate statistical techniques.
```

```python
#Sol7.
data=[1,5,10,15,20]
```

```python
x=min(data)
y=max(data)
```

```python
new=[]
for i in data:
    scaled_data=(i-x)/(y-x)
    new.append(scaled_data)
```
